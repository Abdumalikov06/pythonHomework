```python
file_name = 'data/movie.csv'
df_movies = pd.read_csv(file_name)  
df_movies.columns = df_movies.columns.str.strip()
df_movies["duration"] = pd.to_numeric(df_movies["duration"], errors='coerce')
df_movies["director_facebook_likes"] = pd.to_numeric(df_movies["director_facebook_likes"], errors='coerce')
filtered_movies = df_movies[df_movies["duration"] > 120]
sorted_movies = filtered_movies.sort_values(by="director_facebook_likes", ascending=False)


print(sorted_movies)

```
