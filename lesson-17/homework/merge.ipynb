```python
import sqlite3
import pandas as pd

conn=sqlite3.connect('data/chinook.db')
df1=pd.read_sql_query("Select * from customers", con=conn)
df1
df2=pd.read_sql_query("Select * from invoices",con=conn)
df2
result=pd.merge(df1,df2,on="CustomerId")
result.head()
```

```python
invoice_counts = result.groupby(["CustomerId", "FirstName", "LastName"])["InvoiceId"].count().reset_index(name="Totalinvoice")
invoice_counts.head()
```
