```python
import pandas as pd
```

```python
titanic=pd.read_excel('data/titanic.xlsx')
titanic
result=titanic.groupby("Pclass").agg(
    Average_age=("Age",'mean'),
    Total_fare=("Fare",'sum'),
    Count_passangers=("PassengerId",'count')
).reset_index()
result

```

```python
f=pd.read_csv("data/movie.csv")
f
f.groupby(["color","director_name"]).agg(
    num_critic_for_reviews=("num_critic_for_reviews","sum"),
    avg_duration=("duration","mean")
).reset_index()

```

```python
file=pd.read_parquet("data/flights")
file
result=file.groupby(["Year","Month"]).agg(
    Total=("flights","sum"),
    avg_arrdelay=("ArrDelay","mean"),
    max_Depdelay=("DepDelay","max")

).reset_index()
result
```

```python
import pandas as pd
titanic=pd.read_excel('data/titanic.xlsx')
titanic
def pas_age(x):
    return 'Child' if x>18 else "Adult"

titanic['Age_Group']=titanic['Age'].apply(pas_age)
titanic
```
